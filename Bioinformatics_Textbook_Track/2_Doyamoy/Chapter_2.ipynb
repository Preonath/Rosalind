## TASK 2D: Implement GreedyMotifSearch

```python
def getScore(motifs):
  n = len(motifs)
  m = len(motifs[0])
  score = 0
  for i in range(m):
    val = [0,0,0,0]
    #print(motifs)
    for j in range(n):
      if(motifs[j][i]=='A'):
        val[0] +=1
      elif(motifs[j][i]=='C'):
        val[1] +=1
      elif(motifs[j][i]=='G'):
        val[2] +=1
      elif(motifs[j][i]=='T'):
        val[3] +=1
    mx = 0
    for x in val:
      mx = max(mx,x)
    score +=(n-mx)
  #print(motifs)
  #print(score)
  return score


def getProfile(motifs):
  n = len(motifs)
  m = len(motifs[0])
  profile = []
  lst1 = []
  lst2 = []
  lst3 = []
  lst4 = []
  for i in range(m):
    val = [0.0,0.0,0.0,0.0]
    for j in range(n):
      if(motifs[j][i]=='A'):
        val[0] +=1.0
      elif(motifs[j][i]=='C'):
        val[1] +=1.0
      elif(motifs[j][i]=='G'):
        val[2] +=1.0
      elif(motifs[j][i]=='T'):
        val[3] +=1.0
    lst1.append(float(val[0]/n*1.0))
    lst2.append(float(val[1]/n*1.0))
    lst3.append(float(val[2]/n*1.0))
    lst4.append(float(val[3]/n*1.0))
  profile = [lst1,lst2,lst3,lst4]
  #print(motifs)
  #print(profile)
  return profile


def profileMostProbable(Dna, k, profile):
  ln = len(Dna)
  #print(Dna)
  #print(profile)
  mx = float(0)
  ans = ""
  for i in range(ln-k+1):
    sub = Dna[i:i+k]
    p = 0
    score = float(1.0)
  
    for x in sub:
      if x=='A':
         q = 0
      if x=='C':
        q = 1
      if x=='G':
        q = 2
      if x=='T':
        q = 3
      score = score*float(profile[q][p])
      p = p + 1
    
    if score>=mx:
      mx = score
      ans = sub
  #print(ans)
  return ans


def GREEDYMOTIFSEARCH(dna,k,t):
  bestMotifs = []
  for i in range(0,t,1):
    bestMotifs.append(dna[i][0:k])

  for i in range(0,len(dna[0])-k+1,1):
    motifs = []
    motifs.append(dna[0][i:i+k])
    for j in range(1,t,1):
      profile = getProfile(motifs)
      motifs.append(profileMostProbable(dna[j],k,profile))
    if getScore(motifs) < getScore(bestMotifs):
      bestMotifs = motifs
  return bestMotifs




k = int(input())
t = int(input())
dna = list(map(str,input().split()))

bestmotifs = GREEDYMOTIFSEARCH(dna, k, t)
print(('\n').join(bestmotifs))
```
