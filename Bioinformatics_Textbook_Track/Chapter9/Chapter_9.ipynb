## ***TASK 9H: Pattern Matching with the Suffix Array***

```
def fun(text):
  ln = len(text)
  text = text * 2
  ara = []
  
  for i in range(ln):
    ara.append(text[i:i+ln])
  ara = sorted(ara)
  ans = []
  sz = len(ara)
  for i in range(sz):
    for j in range(sz):
      if ara[i][j] == '$':
        ans.append((sz-1-j,ara[i]))
  return ans[::]


text = input()
text = text + '$'
patterns = list(map(str,input().split()))
ara_text = fun(text)
ans = []
sz = len(ara_text)
for i in range(sz):
  st = ara_text[i][1]
  pos = ara_text[i][0]
  for j in range(len(st)):
    sub = st[j:j+4]
    if sub == patterns:
      ans.append(pos)

ans = set(ans)
print(*ans)
```

## ***TASK 9I: Construct the Burrows-Wheeler Transform of a String***

```
ch = input()
if(ch[-1]!='$'):
  ch = ch+'$'
sz = len(ch)
ch = ch+ch
bwt = []
ans = []

for i in range(sz):
  sub = ch[i:i+sz]
  bwt.append(sub)

bwt.sort()
for x in bwt:
  ans.append(x[-1])
print(('').join(ans))
```

## ***TASK 9J: Reconstruct a String from its Burrows-Wheeler Transform***

```
def Reconstruct_BWT(bwt):

  cnt = dict()
  last_col = []
  
  for x in bwt:
    if x not in cnt:
      cnt[x] = 1
    else:
      cnt[x]+=1
    last_col.append([x,cnt[x]])
  
  
  first_col = last_col[:]
  first_col.sort()
  #print(last_col)
  #print(first_col)
  sz = len(first_col)

  last_to_first = dict()

  for i in range(sz):
    str1 = last_col[i][0]+str(last_col[i][1])
    str2 = first_col[i][0]+str(first_col[i][1])
    last_to_first[str1] = str2

  ans = ""
  ch = '$1'

  while True:
    ch = last_to_first[ch]
    ans+=ch[0]
    if(ch=='$1'):
      break

  
  print(ans)

  #print(last_to_first)


bwt = input()
Reconstruct_BWT(bwt)

```

## ***TASK 9K: Generate the Last-to-First Mapping of a String***

```
def last_to_first(bwt,ind):
  cnt = dict()
  last_col = []
  
  for x in bwt:
    if x not in cnt:
      cnt[x] = 1
    else:
      cnt[x]+=1
    last_col.append([x,cnt[x]])
  
  
  first_col = last_col[:]
  first_col.sort()
  ans = 0
  for x in first_col:
    if(x==last_col[ind]):
      return ans
    ans+=1

bwt = input()  
k = int(input())
print(last_to_first(bwt,k))
```

## ***TASK 9O: Find All Approximate Occurrences of a Collection of Patterns in a String***

```
def HM(str1,str2):
  score = 0
  sz = len(str1)
  for i in range(sz):
    if(str1[i]!=str2[i]):
      score+=1
  return score

dna = input()
patterns = list(map(str,input().split()))
d = int(input())

ans = []
for pattern in patterns:
  k = len(pattern)
  for i in range(len(dna)-k+1):
    sub = dna[i:i+k]
    if(HM(sub,pattern)<=d):
      ans.append(i)

ans.sort()
print(*ans)
```
