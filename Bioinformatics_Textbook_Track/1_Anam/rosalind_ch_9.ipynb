```python
def reconstructStringFromBWT(text):
  last=[]
  first=[]
  count={'A':1,
         'T':1,
         'G':1,
         'C':1}
  ind=0
  cur=0
  for x in text:
    st=''
    if x is not '$':
      st+=x
      st+=str(count[x])
      last.append(st)
      first.append(x)
      count[x]+=1
    else:
      last.append(x)
      first.append(x)
      cur=ind
    ind+=1
  first.sort()
  l=len(text)
  res='$'
  while len(res)<len(text):
    res+=first[cur][0]
    val=first[cur][0]
    temp=cur-first.index(val)+1
    val+=str(temp)
    cur=last.index(val)
    l-=1
  res=res[1:]+'$'
  print(res)
  return 
```

### BA9L: Implement BWMatching

```python
def lastToFirstMapping(text,ind):
  mark={
      'A':1,
      'T':1,
      'G':1,
      'C':1
  }
  first=[]
  last=[]
  BWT=[]
  count=0
  for x in text:
    if x=='$':
      first.append(x)
      last.append(x)
      BWT.append(x)
    else:
      val=x+str(mark[x])
      mark[x]+=1
      first.append(x)
      last.append(val)
      BWT.append(x)
  first.sort()
  temp=last[ind]
  res=first.index(temp[:1])+int(temp[1:])-1
  
  return res

def BWMatching(text,patterns):
  numOfPattern=[]
  BWT=[]
  BWTC=[]
  mark={
      'A':1,
      'T':1,
      'G':1,
      'C':1
  }
  for x in text:
    if x=='$':
      BWT.append(x)
      BWTC.append(x)
    else:
      BWT.append(x)
      BWTC.append(x+str(mark[x]))
      mark[x]+=1
  invBWT=[]
  res=[]
  for pattern in patterns:
    pattern=pattern[::-1]
    top=0
    bottom=len(text)-1
    for x in pattern:
      if x not in BWT[top:bottom+1]:
        top=2
        bottom=1
        break
      topInd=BWT[top:bottom+1].index(x)
      invBWT=BWT[top:bottom+1]
      invBWT.reverse()
      bottomInd=len(invBWT)-invBWT.index(x)-1
      topInd+=top
      bottomInd+=top
      top=lastToFirstMapping(text,topInd)
      bottom=lastToFirstMapping(text,bottomInd)
      if top>bottom:
        break
    res.append(bottom-top+1)  
  for x in res:
    print(x,end=' ')
  return 
```

```python
def suffixArray(text):
  sufArray=[]
  for i in range(len(text)):
    temp=text[i:]
    sufArray.append(temp)        
  sufArray.sort()
  res=sufArray
  sarray=[]
  for x in res:
    sarray.append(len(text)-len(x))
  return sarray
def patternMatchingSArray(sarray,text,pattern):
  minIndex=0
  maxIndex=len(text)-1
  while minIndex<maxIndex:
    midIndex=int((maxIndex+minIndex)/2)
    if (pattern>text[sarray[midIndex]:]):
      minIndex=midIndex+1
    else:
      maxIndex=midIndex
  first=minIndex
  maxIndex=len(text)-1
  while minIndex<maxIndex:
    midIndex=int((maxIndex+minIndex+1)/2)
    if (text[sarray[midIndex]:]>(pattern+'Z')):
      maxIndex=midIndex-1
    else:
      minIndex=midIndex
  last=maxIndex
  if text[sarray[first]:sarray[first]+len(pattern)]!=pattern or text[sarray[last]:sarray[last]+len(pattern)]!=pattern:
    return -1
  return sarray[first:last+1]

def hammingDsitance(t1,t2):
  count=0
  for x,y in zip(t1,t2):
    if x is not y :
      count+=1
  return count+abs(len(t1)-len(t2))
def dividePattern(pattern,d):
  l=int(len(pattern)/d)
  dividedpatterns=[]
  p=0
  for i in range(d):
    if i!=d-1:
      dividedpatterns.append(pattern[p:p+l])
    else:
      dividedpatterns.append(pattern[p:])
    p+=l
  return dividedpatterns

def approximateOccurrences(text,patterns,d):
  sarray=suffixArray(text)
  res=[]
  for pattern in patterns:
    dividedPatterns=dividePattern(pattern,d+1)
    pos=0
    tres={-1}
    for dpattern in dividedPatterns:
      
      temp=patternMatchingSArray(sarray,text,dpattern)
      if temp!=-1:
        for x in temp:
          leftDPattern=pattern[:pos]
          rightDPattern=pattern[pos+len(dpattern):]
          leftText=text[x-len(pattern[:pos]):x]
          rightText=text[x+len(dpattern):x+len(dpattern)+len(pattern[pos+len(dpattern):])]
          if hammingDsitance(leftDPattern,leftText)+hammingDsitance(rightDPattern,rightText)<=d:
            tres.add(x-len(pattern[:pos]))
      pos+=len(dpattern)
    for x in tres:
      if x!=-1:
        res.append(x)

  for x in res:
    print(x,end=' ')
  return      
```
