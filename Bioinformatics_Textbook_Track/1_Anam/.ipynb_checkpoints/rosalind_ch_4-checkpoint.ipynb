### BA4A: Translate an RNA String into an Amino Acid String

```
def constructMapC2A():
  numOfPatOfAcid=[1,3,2,2,4,2,2,2,2,1,1,2,4,2,2,4,4,4,2,2,4,4,4,2,2]
  acidName='MIRSTKNLFW*CS*YVGAEDLRPQH'
  pattern=[]
  aminoAcidMap={}
  text1='AUGC'
  for x in text1:
    text2='UGCA'
    for y in text2:
      text3='GACU'
      for z in text3:
        #print(x+y+z)
        pattern.append(x+y+z)
  cur=-1
  for i in range(len(numOfPatOfAcid)):
    count1=numOfPatOfAcid[i]
    while count1:
      count1-=1
      cur+=1
      #print(pattern[cur]+' '+acidName[i])
      aminoAcidMap[pattern[cur]]=acidName[i]
  return aminoAcidMap   
def rnaToAminoAcid(text):
  aminoAcidMap=constructMapC2A()
  print(aminoAcidMap)
  amino_acid_string=""
  count1=0
  temp=""
  for x in text:
    temp+=x
    if len(temp)==3:
      if aminoAcidMap[temp]!='*':
        amino_acid_string+=aminoAcidMap[temp]
      temp=""
      count1=0
  return amino_acid_string
```

### BA4B: Find Substrings of a Genome Encoding a Given Amino Acid String

```
aminoAcidMap={}
def constructMapA2C():
  numOfPatOfAcid=[1,3,2,2,4,2,2,2,2,1,1,2,4,2,2,4,4,4,2,2,4,4,4,2,2]
  acidName='MIRSTKNLFW*CS*YVGAEDLRPQH'
  pattern=[]
  text1='AUGC'
  for x in text1:
    text2='UGCA'
    for y in text2:
      text3='GACU'
      for z in text3:
        #print(x+y+z)
        pattern.append(x+y+z)
  cur=-1
  for i in range(len(numOfPatOfAcid)):
    count1=numOfPatOfAcid[i]
    aminoAcidMap[acidName[i]]=[]
    while count1:
      count1-=1
      cur+=1
      #print(pattern[cur]+' '+acidName[i])
      #print(acidName[i])
      #print(pattern[cur])

      aminoAcidMap[acidName[i]].append(pattern[cur])
      #print()
  return aminoAcidMap 

def subsequences(dna,l):
  temp=""
  seq=[]
  for x in dna:
    temp+=x
    if len(temp)==l:
      seq.append(temp)
      temp=temp[1:]
  return seq


possibleDnaSeq=[]

def allPossibleString(amino_acid_string,text,i):
  if(i>=len(amino_acid_string)):
    possibleDnaSeq.append(text)
    possibleDnaSeq.append(reverseCompliment(text))
    return
  for x in aminoAcidMap[amino_acid_string[i]]:
    allPossibleString(amino_acid_string,text+rna2dna(x),i+1)
  return 
def reverseCompliment(text):
  re=""
  mark={
      "A":"T",
      "T":"A",
      "G":"C",
      "C":"G"
  }
  for x in text:
    re+=mark[x]
  return re[::-1] 

def rna2dna(text):
  temp=""
  for x in text:
    if x=='U' :
       temp+='T'
    else :
      temp+=x
  return temp

def matchedSeq(text,aaacidSeq):
  constructMapA2C()
  substrings=subsequences(text,(3*len(aaacidSeq)))
  allPossibleString(aaacidSeq,'',0)
  l=1
  res=[]
  substrings.sort()
  possibleDnaSeq.sort()
  for x in substrings :
    for y in possibleDnaSeq:
      if x == y :
        res.append(y)
        print(y)
        l+=1
  
  return    


```

### BA4E: Find a Cyclic Peptide with Theoretical Spectrum Matching an Ideal Spectrum

```
acid2Mass={
      "G":57,
      "A":71,
      "S":87,
      "P":97,
      "V":99,
      "T":101,
      "C":103,
      "I":113,
      "L":113,
      "N":114,
      "D":115,
      "K":128,
      "Q":128,
      "E":129,
      "M":131,
      "H":137,
      "F":147,
      "R":156,
      "Y":163,
      "W":186} 
def Mass(text):
  res=0  
  for x in text:
    res+=acid2Mass[x]
  return res  
    
    
def theoryticalSpectrum(text):
  res=[0]
  temp=text+text
  for i in range(1,len(text)+1):
    for j in range(0,len(text)):
      res.append(Mass(temp[j:j+i]))
      if i==len(text):
        break 
  res.sort()
  return res
def linerSpec(text):
  res=[0]
  temp=text
  for i in range(1,len(text)+1):
    for j in range(0,len(text)):
      if j+i>len(text):
        break
      res.append(Mass(temp[j:j+i]))
      if i==len(text):
        break 
  res.sort()
  return res

def consistent(Peptide,Spectrum):
  tspectrum=linerSpec(Peptide)
  for x in tspectrum:
    if tspectrum.count(x)>Spectrum.count(x):
      return 0
  return 1

def Expand(Peptides,aacid):
  res=[]
  for x in Peptides:
    temp=x
    for y in aacid:
      temp+=y
      res.append(temp)
      temp=x
  return res
def CYCLOPEPTIDESEQUENCING(Spectrum):
  Peptides=[""]
  Spectrum.sort()
  aacid=[]
  mass2Acid={}
  res={""}
  for key,value in acid2Mass.items():
    if value  not in  mass2Acid:
      mass2Acid[value]=[]
    mass2Acid[value].append(key)
  for x in Spectrum:
    if x in mass2Acid:
      for y in mass2Acid[x]:
        aacid.append(y)
  while len(Peptides)!=0:
    temp=Peptides
    Peptides = Expand(Peptides,aacid)
    tempPeptides=Expand(temp,aacid)

    for Peptide in tempPeptides:
      if Mass(Peptide) == Spectrum[-1:][0]:
        if theoryticalSpectrum(Peptide) == Spectrum:
          st=''
          for y in Peptide:
            t=str(acid2Mass[y])+'-'
            st+=t
          res.add(st[:-1])
        Peptides.remove(Peptide)
      elif consistent(Peptide,Spectrum)==0:
        Peptides.remove(Peptide)
  res.discard("")
  for x in res:
    print(x,end=' ')
  return 
```

### BA4F: Compute the Score of a Cyclic Peptide Against a Spectrum

```
def mass(text):
  mark={
      "G":57,
      "A":71,
      "S":87,
      "P":97,
      "V":99,
      "T":101,
      "C":103,
      "I":113,
      "L":113,
      "N":114,
      "D":115,
      "K":128,
      "Q":128,
      "E":129,
      "M":131,
      "H":137,
      "F":147,
      "R":156,
      "Y":163,
      "W":186}
  
  res=0  
  for x in text:
    res+=mark[x]
  return res  
    
    
def Tspectrum(text):
  res=[0]
  temp=text+text
  for i in range(1,len(text)+1):
    for j in range(0,len(text)):
      res.append(mass(temp[j:j+i]))
      if i==len(text):
        break
      
  
  res.sort()
  return res



def score(aminoAcidString,spectrum):
  tspectrum=Tspectrum(aminoAcidString)
  #print(tspectrum)
  uniqueMassInSpectrum={-1}
  for x in tspectrum:
    uniqueMassInSpectrum.add(x)
  score=0
  uniqueMassInSpectrum.remove(-1)
  for x in uniqueMassInSpectrum:
    #print(x)
    #print(tspectrum.count(x))
    score+=min(tspectrum.count(x),spectrum.count(x))
    
  return score  
```

### BA4H: *Generate* the Convolution of a Spectrum

```
def convolution(spectrum):
  res=[]
  store=[]
  unique={-1}
  for x in spectrum:
    for y in spectrum:
      if x-y > 0:
        store.append(x-y)
        unique.add(x-y)
  unique.discard(-1)
  for x in unique:
    res.append((store.count(x),x))
  res.sort()
  res.reverse()
  for x,y  in res:
    for i in range(x):
      print(y,end=' ')
  return
```

```
acid2Mass={
      "G":57,
      "A":71,
      "S":87,
      "P":97,
      "V":99,
      "T":101,
      "C":103,
      "I":113,
      "L":113,
      "N":114,
      "D":115,
      "K":128,
      "Q":128,
      "E":129,
      "M":131,
      "H":137,
      "F":147,
      "R":156,
      "Y":163,
      "W":186} 
def Mass(text):
  res=0  
  for x in text:
    res+=acid2Mass[x]
  return res  
    
    
def theoryticalSpectrum(text):
  res=[0]
  temp=text+text
  for i in range(1,len(text)+1):
    for j in range(0,len(text)):
      res.append(Mass(temp[j:j+i]))
      if i==len(text):
        break 
  res.sort()
  return res
def linerSpec(text):
  res=[0]
  temp=text
  for i in range(1,len(text)+1):
    for j in range(0,len(text)):
      if j+i>len(text):
        break
      res.append(Mass(temp[j:j+i]))
      if i==len(text):
        break 
  res.sort()
  return res
def Score(aminoAcidString,spectrum):
  tspectrum=theoryticalSpectrum(aminoAcidString)
  #print(tspectrum)
  uniqueMassInSpectrum={-1}
  for x in tspectrum:
    uniqueMassInSpectrum.add(x)
  score=0
  uniqueMassInSpectrum.remove(-1)
  for x in uniqueMassInSpectrum:
    #print(x)
    #print(tspectrum.count(x))
    score+=min(tspectrum.count(x),spectrum.count(x))
    
  return score


def Expand(Peptides,aacid):
  res=[]
  for x in Peptides:
    temp=x
    for y in aacid:
      temp+=y
      res.append(temp)
      temp=x
  return res

def Cut(Leaderboard, Spectrum, N):
  map={}
  scores=[]
  res=[]
  for x in Leaderboard:
    sc=Score(x,Spectrum)
    if sc not in map:
      map[sc]=[]
    map[sc].append(x)
    if sc not in scores:
      scores.append(sc)
  scores.sort()
  scores.reverse()
  counti=0
  for x in scores:
    if counti>N:
      break
    for y in map[x]:
      counti+=1
      res.append(y)
  return res

def convolution(M,spectrum):
  res=[]
  store=[]
  unique={-1}
  for x in spectrum:
    for y in spectrum:
      if x-y > 0 and (x-y)<=200:
        store.append(x-y)
        unique.add(x-y)
  unique.discard(-1)
  for x in unique:
    res.append((store.count(x),x))
  res.sort()
  res.reverse()
  counti=1
  output=[]
  pre=0
  print(res)
  for y,x in res:
    if counti>M and pre!=y:
      break
    output.append(x)
    pre=y
    counti+=1
  return output

def ConvolutionCyclopeptideSequencing(M,N,Spectrum):
  Leaderboard=[""]
  LeaderPeptide=""
  Spectrum.sort()
  aacid=[]
  mass2Acid={}
  res=''

  aacidMasses=convolution(M,Spectrum)
  print(aacidMasses)
  for key,value in acid2Mass.items():
    if value  not in  mass2Acid:
      mass2Acid[value]=[]
    mass2Acid[value].append(key)
  for x in aacidMasses:
    if x not in mass2Acid:
      continue
    for y in mass2Acid[x]:
      aacid.append(y)

  while len(Leaderboard)!=0:
    temp=Leaderboard
    Leaderboard = Expand(Leaderboard,aacid)
    tempLeaderboard=Expand(temp,aacid)
    for Peptide in tempLeaderboard:
      if Mass(Peptide) == Spectrum[-1:][0]:
        if Score(Peptide, Spectrum) > Score(LeaderPeptide, Spectrum):
          LeaderPeptide = Peptide
          st=''
          for y in Peptide:
            t=str(acid2Mass[y])+'-'
            st+=t
          res=st[:-1]
        Leaderboard.remove(Peptide)
      elif Mass(Peptide) > Spectrum[-1:][0]:
        Leaderboard.remove(Peptide)
    Leaderboard = Cut(Leaderboard, Spectrum, N)
  print(res)
  return res
```

### BA4L: Trim a Peptide Leaderboard

```
acid2Mass={
      "G":57,
      "A":71,
      "S":87,
      "P":97,
      "V":99,
      "T":101,
      "C":103,
      "I":113,
      "L":113,
      "N":114,
      "D":115,
      "K":128,
      "Q":128,
      "E":129,
      "M":131,
      "H":137,
      "F":147,
      "R":156,
      "Y":163,
      "W":186} 
def Mass(text):
  res=0  
  for x in text:
    res+=acid2Mass[x]
  return res  
    
    
def linerSpec(text):
  res=[0]
  temp=text
  for i in range(1,len(text)+1):
    for j in range(0,len(text)):
      if j+i>len(text):
        break
      res.append(Mass(temp[j:j+i]))
      if i==len(text):
        break 
  res.sort()
  return res

def Score(aminoAcidString,spectrum):
  tspectrum=linerSpec(aminoAcidString)
  #print(tspectrum)
  uniqueMassInSpectrum={-1}
  for x in tspectrum:
    uniqueMassInSpectrum.add(x)
  score=0
  uniqueMassInSpectrum.remove(-1)
  for x in uniqueMassInSpectrum:
    #print(x)
    #print(tspectrum.count(x))
    score+=min(tspectrum.count(x),spectrum.count(x))
    
  return score


def Trim(Leaderboard, Spectrum, N):
  map={}
  scores=[]
  res=[]
  for x in Leaderboard:
    sc=Score(x,Spectrum)
    if sc not in map:
      map[sc]=[]
    map[sc].append(x)
    if sc not in scores:
      scores.append(sc)
  scores.sort()
  scores.reverse()
  counti=0
  for x in scores:
    if counti>N:
      break
    for y in map[x]:
      counti+=1
      res.append(y)
  for x in res:
    print(x,end=' ')
  return 

```
