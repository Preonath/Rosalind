### BA9C: Construct the Suffix Tree of a String

```python
graph={}
def traverse(ind,sym,text):
  if not graph[(ind,sym)] :
    if text!='':
      print(text)
    return
  if sym!='' and len(graph[(ind,sym)])>1:
    print(text)
    text=''
  for (pos,ch) in graph[(ind,sym)]:
    traverse(pos,ch,text+ch)
  return 


def trieConstruct(arr):
  inc=1
  graph[(0,'')]=[]
  nodes={}
  mark={
      'A':0,
      'T':0,
      'G':0,
      'C':0
  }
  for pattern in arr:
    cur=(0,'')
    ind=0
    for ch in pattern:
      #print(cur)
      #print(ind)
      if ind not in nodes:
        nodes[ind]={
                    'A':0,
                    'T':0,
                    'G':0,
                    'C':0,
                    '$':0
                    }
      #print(nodes[ind])
      if nodes[ind][ch]>=1:
        cur=(nodes[ind][ch],ch)
        ind=nodes[ind][ch]
      else:
        nodes[ind][ch]=inc
        graph[cur].append((inc,ch))
        graph[(inc,ch)]=[]
        cur=(inc,ch)
        ind=inc
        inc+=1
      
  return graph,nodes




def suffixTree(text):
  suffixes=[]
  for i in range(len(text)):
    suffixes.append(text[i:])
  suffixes.sort()
  graph,nodes=trieConstruct(suffixes)
  traverse(0,'','')  
  return
f = open("rosalind_ba9c.txt", "r")
#f = open("input.txt", "r")
arr=f.read().split('\n')
f.close()
suffixTree(arr[0])
```
