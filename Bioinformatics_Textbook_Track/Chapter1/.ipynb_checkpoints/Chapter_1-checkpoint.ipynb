##TASK 1C:Find the Reverse Complement of a String

```
def reverse(string):
    res = ""
    for i in string[::-1]:
        if i=='A':
            res += 'T'
        elif i =='T':
            res += 'A'
        elif i=='C':
            res += 'G'
        else:
            res += 'C'
    return res
  
pattern = input("Input the pattern: ")
res = reverse(pattern)
print(res)
```

##TASK 1E:Find Patterns Forming Clumps in a String


```
s = str(input())
k = int(input())
L = int(input())
t = int(input())

ln = len(s)

lst = list()

for i in range(ln-L+1):
  sub_str = s[i:i+L]
  WF = dict()
  
  for j in range(L-k+1):
    pattern = sub_str[j:j+k]
    WF[pattern] = 0
    
  for j in range(L-k+1):
    pattern = sub_str[j:j+k]
    WF[pattern] = WF[pattern] + 1
  
  #print(WF)
  
  for key, value in WF.items():
    if value>=t:
      lst.append(key)

lst = set(lst)
print (' '.join(lst))
```

##TASK 1G:Compute the Hamming Distance Between Two Strings


```
def hmdist(str1,str2):
    dist = 0
    L = len(str1)
    for i in range(L):
        if(str1[i]!=str2[i]):
            dist = dist+1
    return dist
  
str1 = input("Input first string: ")
str2 = input("Input second string: ")
print(hmdist(str1,str2))
```

##TASK 1H:Find All Approximate Occurrences of a Pattern in a String

```
ans = list()
def patternmatch(pattern,text,d):
    L =len(text)
    L2= len(pattern)
    for i in range(L-L2+1):
        cnt = 0
        for j in range(L2):
            if text[i+j]!=pattern[j]:
                cnt+=1
        if cnt<=d:
            ans.append(i)
            
pattern = input("Input the pattern: ")
text = input("Input the text: ")
d = int(input())

patternmatch(pattern,text,d)

print(*ans)
```

##TASK 1I:Find the Most Frequent Words with Mismatches in a String

```
import itertools

def patternmatch(pattern,text,d):
    cnt_match = 0
    L =len(text)
    L2= len(pattern)
    for i in range(L-L2+1):
        cnt = 0
        for j in range(L2):
            if text[i+j]!=pattern[j]:
                cnt+=1
        if cnt<=d:
            cnt_match+=1
    
    return cnt_match

text = input("Input the text: ")
k = int(input())
d = int(input())
L = ['A','C','G','T']
mx = 0
perms = itertools.product(L, repeat = k)

for perm in perms:
    perm_string = ''.join(perm)
    ret = patternmatch(perm_string,text,d)
    if ret>mx:
        mx=ret
ans = list()
perms = itertools.product(L, repeat = k)        
for perm in perms:
    perm_string = ''.join(perm)
    ret = patternmatch(perm_string,text,d)
    if ret == mx:
        ans.append(perm_string)
        
print(*ans)
```

##TASK 1J:Find Frequent Words with Mismatches and Reverse Complements

```
def patternmatch(pattern,text,d):
    cnt_match = 0
    L =len(text)
    L2= len(pattern)
    for i in range(L-L2+1):
        cnt = 0
        for j in range(L2):
            if text[i+j]!=pattern[j]:
                cnt+=1
        if cnt<=d:
            cnt_match+=1
    
    return cnt_match

text = input("Input the text: ")
k = int(input())
d = int(input())
L = ['A','C','G','T']
mx = 0
perms = itertools.product(L, repeat = k)

for perm in perms:
    perm_string = ''.join(perm)
    ret = patternmatch(perm_string,text,d)
    rev_string = reverse(perm_string)
    ret2 = patternmatch(rev_string,text,d)
    if ret+ret2>mx:
        mx=ret+ret2
        
ans = list()
perms = itertools.product(L, repeat = k)        
for perm in perms:
    perm_string = ''.join(perm)
    ret = patternmatch(perm_string,text,d)
    rev_string = reverse(perm_string)
    ret2 = patternmatch(rev_string,text,d)
    if ret+ret2 == mx:
        ans.append(perm_string)
        
print(*ans)
```

##TASK 1K:Generate the Frequency Array of a String

```
dna = input()
sz = len(dna)
k = int(input())
L = ['A','C','G','T']
perms = itertools.product(L, repeat = k)
ans = []
for perm in perms:
    perm_string = ''.join(perm)
    cnt = 0
    for i in range(sz+k-1):
      sub = dna[i:i+k]
      if(sub==perm_string):
        cnt = cnt+1
    
    ans.append(cnt)
    
print(*ans)
```

##TASK 1M:Implement NumberToPattern

```
def NumberToPattern(index, k):
    L = ['A', 'C', 'G', 'T']
    
    perms = itertools.product(L, repeat = k)
    all_string = []
    for perm in perms:
        perm_string = ''.join(perm)
        all_string += [perm_string]
    return all_string[index]
  
index = int(input("input the index: "))
k = int(input())

print(NumberToPattern(index,k))
```

##TASK 1N:Generate the d-Neighborhood of a String

```
def hmdist(str1,str2):
    dist = 0
    L = len(str1)
    for i in range(L):
        if(str1[i]!=str2[i]):
            dist = dist+1
    return dist
  
  
pattern = input()
k = len(pattern)
d = int(input())

L = ['A', 'C', 'G', 'T']
perms = itertools.product(L, repeat = k)
ans = []
for perm in perms:
    perm_string = ''.join(perm)
    h = hmdist(perm_string,pattern)
    if(h<=d):
      ans.append(perm_string)
ans = set(ans)      
print(('\n').join(ans))
```
