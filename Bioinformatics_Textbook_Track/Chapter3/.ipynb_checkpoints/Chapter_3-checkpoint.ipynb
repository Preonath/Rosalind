## TASK 3A: Generate the k-mer Composition of a String

```
k = int(input())

s = input()

ln = len(s)
kmer = []
for i in range(ln-k+1):
  sub = s[i:i+k]
  kmer.append(sub)
  
kmer.sort()

print(("\n").join(kmer))
```

## TASK 3B: Reconstruct a String from its Genome Path

```
ans = ""

f = 1
dna = list(map(str,input().split()))
for read in dna:
  if(f==1):
    ans = ans + read
    f = 0
  else:
    ln = len(read)
    ans = ans + read[ln-1]
  
  
print(ans)
```

## TASK 3C : Construct the Overlap Graph of a Collection of k-mers

```
cnt = 0
ln = 0


lst = list(map(str,input().split()))
cnt = len(lst)   
lst.sort()
ln = len(lst[0])
for i in range(cnt):
  for j in range(cnt):
    if(i==j):
      continue
    sub1 = lst[i][1:ln]
    sub2 = lst[j][0:ln-1]
    if(sub1==sub2):
      print(lst[i]+" -> "+lst[j])
```

## TASK 3D: Construct the De Bruijn Graph of a String

```

def suffix(str):
  return str[:len(str)-1]

def prefix(str):
  return str[1:]


  
k = int(input())
dna = input()
size = len(dna)
kmers = []
for i in range(size-k+1):
  sub = dna[i:i+k]
  kmers.append(sub)
kmers.sort()
len_kmers = len(kmers)


ind = 0
prefix = ""
res = []
while(ind<len_kmers):
  
  newprefix = kmers[ind][:k-1]
  if(newprefix==prefix):
    res.append(kmers[ind][1:])            
  else:
    res = ','.join(res)
    if(prefix!=""):
      print(prefix+ " -> "+res)
    res = []
    res.append(kmers[ind][1:])  
    prefix = newprefix
  
  ind +=1
               
    
res = ','.join(res)
print(prefix+ " -> "+res)         
  
  
#print(kmers)
```

## TASK 3E: Construct the De Bruijn Graph of a Collection of k-mers

```
kmers = list(map(str,input().split()))
kmers.sort()
len_kmers = len(kmers)
k = len(kmers[0])

ind = 0
prefix = ""
res = []
while(ind<len_kmers):
  
  newprefix = kmers[ind][:k-1]
  if(newprefix==prefix):
    res.append(kmers[ind][1:])            
  else:
    res = ','.join(res)
    if(prefix!=""):
      print(prefix+ " -> "+res)
    res = []
    res.append(kmers[ind][1:])  
    prefix = newprefix
  
  ind +=1
               
    
res = ','.join(res)
print(prefix+ " -> "+res)         
  
  
#print(kmers)
```

## TASK 3F: Find an Eulerian Cycle in a Graph

```
# 0 -> 3
# 1 -> 0
# 2 -> 1,6
# 3 -> 2
# 4 -> 2
# 5 -> 4
# 6 -> 5,8
# 7 -> 9
# 8 -> 7
# 9 -> 6
from collections import defaultdict 
import numpy as np
import re


graph = defaultdict(list) 
def addEdge(graph,u,v): 
    graph[u].append(v)

circuit = []
vis = np.zeros(10000000)

def heirholzer(u):
  if(vis[u]):
    circuit.append(u)
    return
  print(u)
  vis[u] = 1
  for v in graph[u]:
    heirholzer(v)
  
  circuit.append(u)
  return 
while(1):
  s = input()
  line =  list(map(int, re.findall('\d+', s)))
  sz = len(line)
  if(sz==0):
    break
  for i in range(1,sz):
      addEdge(graph,line[0],line[i])
      
heirholzer(0)
out = ""
for i in range(len(circuit)):
  if(i==0):
    out+=str(circuit[i])
  else:
    out+="->"+str(circuit[i])
    
print(out)
```
