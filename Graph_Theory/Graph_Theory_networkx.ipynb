## **Creating a graph** ##
### **Create an empty graph with no nodes and no edges.** ###

```python
import networkx as nx
G = nx.Graph()
```

> By definition, a Graph is a collection of nodes (vertices) along with identified pairs of nodes (called edges, links, etc). 

> In NetworkX, nodes can be any hashable object e.g., a text string, an image, an XML object, another Graph, a customized node object, etc.

### **Nodes** ### 
 > The graph G can be grown in several ways. 
 
 > NetworkX includes many graph generator functions and facilities to read and write graphs in many formats. 
 
 > To get started though weâ€™ll look at simple manipulations. You can add one node at a time,

```python
G.add_nodes_from([2,3])
nx.draw(G,with_labels=True, font_weight='bold')
```

> You can also add nodes along with node attributes if your container yields 2-tuples of the form (node, node_attribute_dict):

```python
G.add_nodes_from([
    (4, {"color": "red"}),
    (5, {"color": "green"}),
])
nx.draw(G,with_labels=True, font_weight='bold')
```

 > Nodes from one graph can be incorporated into another:



```python
H = nx.path_graph(10)
G.add_nodes_from(H)
nx.draw(G,with_labels=True, font_weight='bold')
```

## **Edges** ##

 > G can also be grown by adding one edge at a time,



```python
G.add_edge(1, 2)
nx.draw(G,with_labels=True, font_weight='bold')
```

```python
e = (2, 3)
G.add_edge(*e)  # unpack edge tuple*
```

```python
nx.draw(G,with_labels=True, font_weight='bold')
```

by adding a list of edges,



```python
G.add_edges_from([(1, 2), (1, 3)])
```

```python
nx.draw(G,with_labels=True, font_weight='bold')
```

```python
import networkx as nx

# Create an example graph (you can replace this with your own graph)
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4), (4, 5), (5, 6)])

# Perform BFS starting from a given node
start_node = 1
bfs_result = list(nx.bfs_edges(G, start_node))

# Print the BFS traversal result
print("BFS traversal starting from node {}: {}".format(start_node, bfs_result))
```
