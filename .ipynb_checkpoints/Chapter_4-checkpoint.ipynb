## ***TASK 4B: Find Substrings of a Genome Encoding a Given Amino Acid String***

```
def rev_com(string):
  dic = {'A':'U','U':'A','G':'C','C':'G'}
  ret = ""
  for x in string:
    ret+=dic[x]
    
  return ret

dna = input()

dna = dna.replace('T','U')

pattern = input()
len_pattern = len(pattern)
len_dna = len(dna)

k = len_dna-len_pattern*3+1
ans = []
for i in range(k):
  ln = len_pattern*3
  sub = dna[i:i+ln]
  rev = rev_com(sub[::-1])
  
  #print(sub + " " + rev)
  s1 = ""
  s2 = ""
  for j in range(0,ln,3):
    s1+=dic[sub[j:j+3]]
    s2+=dic[rev[j:j+3]]
    
  if(s1==pattern or s2==pattern):
    ans.append(sub.replace('U','T'))
    

print(("\n").join(ans))
```

## ***TASK 4C: Generate the Theoretical Spectrum of a Cyclic Peptide***

```
mp = {
    'G':57,
    'A':71,
    'S':87,
    'P':97,
    'V':99,
    'T':101,
    'C':103,
    'I':113,
    'L':113,
    'N':114,
    'D':115,
    'K':128,
    'Q':128,
    'E':129,
    'M':131,
    'H':137,
    'F':147,
    'R':156,
    'Y':163,
    'W':186
}

def seq_to_val(sq):
  val = 0
  for x in sq:
    val+=mp[x]
    
  return val

pro= input()
sz = len(pro)
tmp = pro + pro
ans = []
ans.append(0)
for i in range(1,sz):
  for j in range(sz):
    sub = tmp[j:j+i]
    ret = seq_to_val(sub)
    ans.append(ret)

ans.append(seq_to_val(pro))
ans.sort()

print(*ans)
```

## ***TASK 4D: Compute the Number of Peptides of Given Total Mass***

```
dp = dict()
arr = [57,71,87,97,99,101,103,113,114,115,128,129,131,137,147,156,163,186]
sz = len(arr)
def rec(x):
  if(x<0):
    return 0
  elif(x==0):
    return 1
  elif x in dp:
    return dp[x]
  ret = 0
  for val in arr:
    ret = ret + rec(x-val)
  dp[x] = ret
  return ret

wgt = int(input())

print(rec(wgt))
```

## ***TASK 4E: Find a Cyclic Peptide with Theoretical Spectrum Matching an Ideal Spectrum***

```
mp = {
    'G':57,
    'A':71,
    'S':87,
    'P':97,
    'V':99,
    'T':101,
    'C':103,
    'I':113,
    'L':113,
    'N':114,
    'D':115,
    'K':128,
    'Q':128,
    'E':129,
    'M':131,
    'H':137,
    'F':147,
    'R':156,
    'Y':163,
    'W':186
}

mp2 = {
    'G':57,
    'A':71,
    'S':87,
    'P':97,
    'V':99,
    'T':101,
    'C':103,
    'I':113,
    'N':114,
    'D':115,
    'K':128,
    'E':129,
    'M':131,
    'H':137,
    'F':147,
    'R':156,
    'Y':163,
    'W':186
}


def Mass(Peptide):
  mass = 0
  for x in Peptide:
    mass+=mp[x]
  return mass

def Cyclospectrum(pro):
  sz = len(pro)
  tmp = pro + pro
  ans = []
  ans.append(0)
  for i in range(1,sz):
    for j in range(sz):
      sub = tmp[j:j+i]
      ret = Mass(sub)
      ans.append(ret)

  ans.append(Mass(pro))
  ans.sort()
  return ans



def ParentMass(Peptide):
  Peptide.sort()
  return Peptide[-1]

def Expand(Peptides):
  New_Peptides = []
  for peptide in Peptides:
    for key,value in mp.items():
      New_Peptides.append(peptide+key)

  return New_Peptides

def convert(res):
  ans = []
  for peptide in res:
    tmp = []
    for x in peptide:
      tmp.append(str(mp[x]))
    ans.append(('-').join(tmp))
  
  return ans[::-1]

def LinearSpectrum(peptide):
  sz = len(peptide)
  tmp = peptide
  ans = []
  ans.append(0)
  for i in range(1,sz):
    for j in range(sz):
      if(j+i>sz):
        break
      sub = tmp[j:j+i]
      ret = Mass(sub)
      ans.append(ret)

  ans.append(Mass(peptide))
  ans.sort()
  return ans



def consistent(Peptide,Spectrum):
  tmp = Spectrum[:]
  linear_spectra = LinearSpectrum(Peptide)
  for val in linear_spectra:
    if not val in tmp:
      return False
    else:
      tmp.remove(val)
  return True

def CYCLOPEPTIDESEQUENCING(Spectrum):
  ans = []
  Peptides = [""]
  while len(Peptides)!=0:
    Peptides = Expand(Peptides)
    #print(Peptides)
    New_peptides = Peptides
    for Peptide in Peptides:
      if Mass(Peptide) == ParentMass(Spectrum):
        if Cyclospectrum(Peptide) == Spectrum:
          # print(Peptide)
          ans.append(Peptide)
        New_peptides.remove(Peptide)
      elif not consistent(Peptide,Spectrum):
        New_peptides.remove(Peptide)

    Peptides = New_peptides
    print(Peptides)
    n = len(Peptides[0])
    print(n)
    n+=1
    if(n*(n-1)+2>len(Spectrum)):
      break

  ans = convert(ans)
  #ans = set(ans)
  ans = list(ans)
  ans.sort()
  ans = ans[::-1]
  print(*ans)

Spectrum = list(map(int,input().split()))
CYCLOPEPTIDESEQUENCING(Spectrum)
```

## ***TASK 4G: Implement LeaderboardCyclopeptideSequencing***

```
mp = {
    'G':57,
    'A':71,
    'S':87,
    'P':97,
    'V':99,
    'T':101,
    'C':103,
    'I':113,
    'L':113,
    'N':114,
    'D':115,
    'K':128,
    'Q':128,
    'E':129,
    'M':131,
    'H':137,
    'F':147,
    'R':156,
    'Y':163,
    'W':186
    }

def Mass(Peptides):
  mass = 0
  for x in Peptides:
    mass+=mp[x]
  return mass


def ParentMass(Spectrum):
  Spectrum.sort()
  return Spectrum[-1]

def Expand(Peptides1):
  New_Peptides = []
  for peptide in Peptides1:
    for key,value in mp.items():
      New_Peptides.append(peptide+key)

  return New_Peptides

def spectrum_calc(pro):

  def seq_to_val(sq):
    val = 0
    for x in sq:
      val+=mp[x]
    
    return val

   
  sz = len(pro)
  tmp = pro 
  ans = []
  ans.append(0)
  for i in range(1,sz):
    for j in range(sz):
      if(j+i>sz):
        break
      sub = tmp[j:j+i]
      ret = seq_to_val(sub)
      ans.append(ret)

  ans.append(seq_to_val(pro))
  ans.sort()

  return ans

def Score(theo_spectrum, exp_spectrum):
  score = 0
  tmp = exp_spectrum[:]
  for item in theo_spectrum:
    if item in tmp:
      score+=1
      tmp.remove(item)
  
  return score



def Trim(Leaderboard, Spectrum, N):
  score_list = []
  for peptide in Leaderboard:
    theo_spec = spectrum_calc(peptide)
  
    score = Score(theo_spec,Spectrum)
    score_list.append((score,peptide))

  score_list.sort(reverse=True)




  ans = []
  sc = 0
  for i in range(N):
    if(i>=len(score_list)):
      break
    ans.append(score_list[i][1])
    if(i==N-1):
      ind = i+1
      while(ind<len(score_list) and score_list[ind][0]==score_list[i][0]):
        ans.append(score_list[ind][1])
        ind+=1

  return ans



def LEADERBOARDCYCLOPEPTIDESEQUENCING(Spectrum, N):
  Leaderboard = [""]
  LeaderPeptide = ""
  while len(Leaderboard)>0:
    Leaderboard1 = Expand(Leaderboard)
    for Peptide in Leaderboard1:
      if Mass(Peptide) == ParentMass(Spectrum):
        if Score(Peptide, Spectrum) > Score(LeaderPeptide, Spectrum):
          LeaderPeptide = Peptide
        elif Mass(Peptide) > ParentMass(Spectrum):
          Leaderboard.remove(Peptide)
      Leaderboard = Trim(Leaderboard, Spectrum, N)
  print(LeaderPeptide)

N = int(input())
Spectrum = list(map(int,input().split()))

LEADERBOARDCYCLOPEPTIDESEQUENCING(Spectrum, N)

```

## ***TASK 4H: Generate the Convolution of a Spectrum***

```
ls = list(map(int, input().split()))
res = []
for r in ls:
    for c in ls:
        if(r-c>0):
            res.append(r-c)
            
mp = dict()

for x in res:
    if x not in mp:
        mp[x] = 1
    else:
        mp[x] +=1

#print(mp)


mp = sorted(mp.items(), key=lambda kv: -1*kv[1])

ans = []
for x,y in mp:
    for i in range(y):
        ans.append(x)

print(*ans)
```
